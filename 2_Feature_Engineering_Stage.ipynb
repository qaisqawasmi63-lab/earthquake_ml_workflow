```python
import pandas as pd

df = pd.read_csv("earthquake_preprocessed.csv", parse_dates=['Datetime'])
print(df.head())
```

```python
# Extract time-based features
df['Year'] = df['Datetime'].dt.year
df['Month'] = df['Datetime'].dt.month
df['DayOfWeek'] = df['Datetime'].dt.dayofweek
```

```python
# 30-day rolling average magnitude - captures recent regional seismic activity
df = df.set_index('Datetime')
df['RollingMag_30D'] = df['Magnitude'].rolling('30D').mean()
df = df.reset_index()
df['RollingMag_30D'] = df['RollingMag_30D'].fillna(df['Magnitude'].median())
```

```python
# Bin Magnitude into categories for the classification task
bins = [0, 5.9, 6.9, 10]
labels = ['Minor', 'Moderate', 'Major']
df['MagCategory'] = pd.cut(df['Magnitude'], bins=bins, labels=labels)
print(df['MagCategory'].value_counts())
```

```python
engineered_file = "earthquake_engineered.csv"
df.to_csv(engineered_file, index=False)
print(f"Feature engineered data saved to {engineered_file}")
```
