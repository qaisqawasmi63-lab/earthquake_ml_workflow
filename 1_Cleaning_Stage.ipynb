```python
import pandas as pd

# Loading the raw earthquake dataset
df = pd.read_csv("earthquake.csv")
print("-Original Data Info")
df.info()
```

```python
# Keep only actual earthquakes (drop nuclear explosions, rock bursts - different physical events)
df = df[df['Type'] == 'Earthquake'].copy()
print(f"Rows after filtering to Type == 'Earthquake': {len(df)}")
```

```python
# Combine Date + Time into a single Datetime column
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], errors='coerce')
df = df.drop(columns=['Date', 'Time'])
```

```python
# Drop columns that are more than 50% missing - not reliable to impute
missing_ratio = df.isnull().mean()
high_missing_cols = missing_ratio[missing_ratio > 0.5].index.tolist()
print("Dropping high-missing columns:", high_missing_cols)
df = df.drop(columns=high_missing_cols)
```

```python
# Drop rows missing essential fields, sort by time
df = df.dropna(subset=['Latitude', 'Longitude', 'Depth', 'Magnitude', 'Datetime'])
df = df.sort_values('Datetime').reset_index(drop=True)
print("\n-Cleaned Data Info")
df.info()
```

```python
processed_file = "earthquake_preprocessed.csv"
df.to_csv(processed_file, index=False)
print(f"Processed data saved to {processed_file}")
```
